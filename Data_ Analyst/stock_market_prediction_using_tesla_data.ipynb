```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tesla = pd.read_csv('/content/tesla.csv')
tesla.head()
```

```python
tesla.info()
```

```python
tesla['Date'] = pd.to_datetime(tesla['Date'])
tesla['Date'].head()
```

```python
print(f'Dataframe contains stock prices between {tesla.Date.min()} {tesla.Date.max()}')
print(f'Total days = {(tesla.Date.max()  - tesla.Date.min()).days} days')
```

```python
tesla.describe()
```

```python
tesla[['Open','High','Low','Close','Adj Close']].plot(kind='box')
```

```python
plt.plot(tesla['Date'],tesla['Close'])
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Stock Prices of Tesla')
```

```python
# Building the regression model
from sklearn.model_selection import train_test_split

#For preprocessing
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

#For model evaluation
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import r2_score
```

```python
#Split the data into train and test sets
X = np.array(tesla.index).reshape(-1,1)
Y = tesla['Close']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=101)
```

```python
# Feature scaling
scaler = StandardScaler().fit(X_train)
```

```python
from sklearn.linear_model import LinearRegression
#Creating a linear model
lm = LinearRegression()
lm.fit(X_train, Y_train)
```

```python
plt.scatter(X_train, Y_train, label='Actual')
plt.plot(X_train, lm.predict(X_train), color='red', label='Predicted')
plt.legend()
plt.title('Stock Prices of Tesla')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
#Calculate scores for model evaluation
scores = f'''
{'Metric'.ljust(5)}{'Train'.center(20)}{'Test'.center(20)}
{'r2_score'.ljust(10)}{r2_score(Y_train, lm.predict(X_train))}\t{r2_score(Y_test, lm.predict(X_test))}
{'MSE'.ljust(8)}{mse(Y_train, lm.predict(X_train))}\t{mse(Y_test, lm.predict(X_test))}
'''
print(scores)
```
